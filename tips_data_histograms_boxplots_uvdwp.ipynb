```python
import seaborn as sns # for plotting
import matplotlib.pyplot as plt # for showing plots

# load in the data set
tips_data = sns.load_dataset("tips")
```

```python
tips_data.describe()
```

**Creating a Hitogram**

```python
# plot a histogram of total bill
# kde=False --> coz don't want to display density plot
sns.distplot(tips_data["total_bill"], kde=False).set_title("Histogram of Total Bill")

plt.show()
```

```python
# plot a histogram of the tip
# kde=True --> by default. 
sns.distplot(tips_data["tip"]).set_title("Histogram of Total Tip")

plt.show()
```

```python
# Plot a histogram of both - not recommended with homogeneous data coz won't be able to discern which one we are looking at
sns.distplot(tips_data["total_bill"], kde=False)
sns.distplot(tips_data['tip'], kde=False).set_title("Histogram of Both Total Bill and Tip Size")

plt.show()

# good info abt shape but not abt central tendency
```

**Creating a Boxplot**

Boxplots do not show the shape of the distribution, but they can give us a better idea about the center and spread of the distribution as well as any potential outliers that may exist. Boxplots and histograms often complement each other and help the analyst get more information about the data.

```python
# Create a boxplot of the total bill
import seaborn as sns
sns.boxplot(tips_data["total_bill"]).set_title("Box Plot of The Total Bill")
plt.show()
```

```python
# Create a boxplot of the total tip
import seaborn as sns
sns.boxplot(tips_data['tip']).set_title("Box Plot of The Total Tip")
plt.show()
```

```python
# Create a box plot of both total bill and tip

sns.boxplot(tips_data["total_bill"])
sns.boxplot(tips_data['tip']).set_title("Box plot of the total bill and tip")

plt.show()
```

**Creating Histograms and Boxplots Plotted by Groups**

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Using graphs, we can see if there is a difference between the tipping amounts of smokers vs non-smokers. If tipping varies according to the time of the day, or we can explore other trends in the data as well.

```python
# Plotting tips grouped by smoking

sns.boxplot(x=tips_data['tip'], y=tips_data['smoker'])

g = sns.FacetGrid(tips_data, row='smoker')

g = g.map(plt.hist, 'tip')

plt.show()
```

```python
# Plotting tips grouped by time of day

sns.boxplot(x=tips_data['tip'], y=tips_data['time'])

g = sns.FacetGrid(tips_data, row='time')
g = g.map(plt.hist, 'tip')

plt.show()
```

```python
# Plotting tips grouped by day of the week

sns.boxplot(x = tips_data['tip'], y=tips_data['day']).set_title("Box plot of tips by day of the week")

g = sns.FacetGrid(tips_data, row='day')
g = g.map(plt.hist, 'tip')

plt.show()
```
