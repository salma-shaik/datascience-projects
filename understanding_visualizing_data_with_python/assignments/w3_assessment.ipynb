## Scatter plots

First, let's looks at two variables that we expect to have a strong relationship, 'Height' and 'Wingspan'.

````python
# Make a Seaborn scatter plot with x = height and y = wingspan using sns.scatterplot(x, y)
sns.regplot(x='Height', y='Wingspan', data=df)
plt.show()
````

How would you describe the relationship between 'Height' and 'Wingspan'?   
Questions you can ask:
* Is it linear?
* Are there outliers?
* Are their ranges similar or different?  

How else could you describe the relationship?

_The scatter plot presents a positive linear relationship height and wingspan. There is a strong correlation between the variables. Data seems to be pretty normally distributed and not much variability as most of the points are lying on and around the best fit line. There are no major obvious outliers._

````python
sns.lmplot(x='Wingspan', y='Height', hue='Gender', data=df)
plt.show()
````

Now let's look at two variables that we don't yet assume have a strong relationship, 'Wingspan' and 'CWDistance'

````python
iqr_cwd = stats.iqr(df.CWDistance)
iqr_ws = stats.iqr(df.Wingspan)

print(iqr_cwd, iqr_ws)
````

## Barcharts
Now lets plot barplots of 'Glasses'

````python
# Make a Seaborn barplot with x = glasses and y = cartwheel distance
sns.barplot(x='Glasses', y='CWDistance', data=df)
plt.show()
````

What can you say about the relationship of 'Glasses' and 'CWDistance'?

_Seems like Glasses don't play a major role in the cartwheel distance. Both have almost same number of values and similar maximum value around 80._

````python
# Make the same Seaborn boxplot as above, but include gender for the hue argument
sns.barplot(x='Glasses', y='CWDistance', hue='Gender', data=df)
plt.show()
````

How does this new plot change your interpretation about the relationship of 'Glasses' and 'CWDistance'?

_Among people who have glasses, men had higher cartwheel distance than women overall. Whereas in the group without glasses, females have the higher cartwheel distance overall._
